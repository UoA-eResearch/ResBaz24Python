- The word integer comes from the Latin integer meaning "whole" or (literally) "untouched", from in ("not") plus tangere ("to touch"). 
- "Entire" derives from the same origin via the French word entier, which means both entire and integer
- We can use Python as a calculator


````python
2+2 
````

If we were not to use Colab or another kind of Notebook where we interactively see the output, but maybe run our code on a command-line interface, we would use the `print()` function to display the output:

````python
print(2+2)
````

Maybe we want to add more context (more textual description) to the output, so we can add to the `print`


````python
print("The result of 2+2 is: ") 
2+2
````

What we did here:
- we used the `print` function (we can spot that it is a function because of the parentheses) `()`
- the text we put in this (i.e., *The result of 2+2 is:*) is a `string`which is yet another file type in Python
  - side-note: Python doesn't use hte `char` file type to efficiently capture a single character, but also uses `string` to capture a single character
- we had to use the Quotation marks (aka Quotes, Speech marks,...) `"`
  - we could have used single quotes `'` instead of double quotes (aka Apostrophe, Single quotes/quotation marks) but to avoid confusion with genitive `'s` in English aka *possessive apostrophe*
- here we used a workaround and added a new line
  - alternatively, we could used the `+` operator to concatenate (join) the strings together

We could also write that in one line, but that requires us to *cast* the 2+2 to a string, so we can concatenate it with the other string.

````python
print("The result of 2+2 is: " + str(2+2))
````
