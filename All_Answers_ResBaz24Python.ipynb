# Option A: The Math problem üßÆ

### A1: Using a function to create a distribution and fit a linear regression with matplotlib and scipy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate 100 random x values
x_values = np.arange(100)

# Generate random perturbations (delta) for each data point
delta = np.random.uniform(-10, 10, size=(100,))

# Create y values using the linear function y = 0.4*x + 3 + delta
def f(x):
    return 0.4 * x + 3 + delta

y_values = f(x_values)
```

```python
# Plot the scatterplot
plt.scatter(x_values, y_values, label="Data points", color="blue")
```

```python
# Calculate the linear regression
slope, intercept, r_value, p_value, stderr = stats.linregress(x_values, y_values)

# Create the regression line
regression_line = slope * x_values + intercept

# Plot the scatterplot
plt.scatter(x_values, y_values, label="Data points", color="blue")
plt.plot(x_values, regression_line, label="Regression line", color="red")

# Add labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatterplot with Linear Regression Line")

# Show the plot
plt.legend()
plt.show()
```

### A2: Fitting and plotting a non-linear distribution with sklearn and seaborn and pandas

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Define the function
def f(x): # we have created a function simlar to our 'school maths' notation; we could have given it any name
    return x**5 # this is the Python notation for 'to the power of 5'

# Generate values for x
x = np.linspace(-2, 2, 50)  # this is not working with continuous values, but with a range of 50 values between -2 and 2

# Compute corresponding y values
y = f(x)


```

#### Plotting: Using the Seaborn package
We can also use another package to make the plot more appealing. We will use seaborn for this.

```python
# Plot the function using Seaborn
sns.lineplot(x=x, y=y)
```

#### Seaborn: A better workflow
- we create a Pandas DataFrame
  - think of this as a (fancy/powerful) spreadsheet in Python
- we give the columns proper names
- then we don't have to use workaround for axis title



```python
import pandas as pd
# Create a DataFrame
A2data = pd.DataFrame({"xSteps": x, "yClean": y})
A2data.head(10)
```

```python
# Plot the function using Seaborn
sns.lineplot(data=A2data, x="xSteps", y="yClean", legend="full")
```

#### We generate some spreadsheet data with our noise definition

- we will use a [Gaussian/Normal distribution](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) to add some noise to our data
- while we are at it, we also plot the new data to compare it

A2.2: Let's create some noisy data and store it as a spreadsheet and a Pandas Dataframe

```python
# Add Gaussian noise
mean = 0
std_dev = 4  # Adjust the standard deviation to control the noise level
noise = np.random.normal(mean, std_dev, A2data['yClean'].shape)

# Create the noisy y values based on the existing yClean column
A2data['yNoisy'] = A2data['yClean'] + noise

# Display the top 10 lines of the DataFrame
print(A2data.head(10))

# Savining a pandas dataframe to csv
A2data.to_csv('A2data.csv', index=False)
```

#### A2.3: Let's plot the noisy data

```python
# Plotting
sns.scatterplot(data=A2data, x="xSteps", y="yNoisy", label="Noisy Data", color="red")
plt.title("Adding Gaussian Noise to y = x^5")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### A2.3 Let's start with the regression (linear and polynomial)

- there are various ways of doing this
- there are diffferent packages to be used
- some are better for visualisation
- some offer detailed statistical analysis
  - `numpy` is often used, it provides some regression functions, often it is called 'under the hood'
  - `scipy` is a more advanced package, it offers more statistical analysis
  - `sklearn` is a machine learning package, it offers a lot of different models, including regression models
  - `statsmodels` is a package that is often used in academia, it offers a lot of statistical analysis

For this session, we will use `sklearn` as it gives us a good balance between visualisation and statistical analysis

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


# Extracting features (X) and target variable (y)
X = A2data[['xSteps']]
y = A2data['yNoisy']

# Transforming features to polynomial features of degree 3
poly = PolynomialFeatures(degree=5, include_bias=False)
X_poly = poly.fit_transform(X)

# Fitting the model
model = LinearRegression()
model.fit(X_poly, y)

X_fit_poly = poly.transform(X)
y_fit = model.predict(X_fit_poly)

# plot distribution
sns.scatterplot(data=A2data, x="xSteps", y="yNoisy", label="Distribution", color="green")

# Plotting the polynomial regression curve
plt.plot(X, y_fit, color='red', label='Polynomial Regression Curve')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()

# Print the coefficients
print("Coefficients:", model.coef_)
```

Let's have a look at the non-number data types
- Don't worry, this looks like 'a lot' but, in essence, it is a loop over all the categorials (so not numeric data types) and then we plot them

```python
# Create a figure with subplots
fig, axes = plt.subplots(nrows=len(tips.select_dtypes(include='object').columns), ncols=1, figsize=(8, 4 * len(tips.select_dtypes(include='object').columns)))

# Loop over each categorical column and plot
for ax, column in zip(axes, tips.select_dtypes(include='object').columns):
    tips[column].value_counts().plot(kind='bar', ax=ax)
    value_counts = tips[column].value_counts()
    bars = value_counts.plot(kind='bar', ax=ax)

plt.tight_layout(h_pad=5.0)
plt.show()
```
